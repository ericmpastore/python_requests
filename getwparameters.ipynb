```python
searchterm = input("Enter a search term: ")
```

```python
url = "https://www.youtube.com/search"
```

```python
PARAMS = {'q' : searchterm}
```

```python
r_get = requests.get(url=url,params=PARAMS)
```

```python
print(r_get.status_code)
```

```python
r_get.url
```

```python
webbrowser.open(r_get.url)
```
