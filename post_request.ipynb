```python
import requests
from pprint import pprint
import webbrowser
```

```python
headers = {
    'User-Agent': 'MediaWiki REST API docs examples/0.1 (https://www.mediawiki.org/wiki/API_talk:REST_API)'
}

r_post = requests.post('https://en.wikipedia.org/wiki/Main_Page',data={'search':'Skillsoft'},headers=headers)

```

```python
r_post.status_code
```

```python
type(r_post)
```

```python
url = r_post.text
# webbrowser.open(url)
pprint(url)
```

```python
with open("skillsoft.html", "wb") as f:
    for chunk in r_post.iter_content(chunk_size=10000):
        f.write(chunk)
```

```python
webbrowser.open("skillsoft.html")
```
