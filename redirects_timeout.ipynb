```python
import requests
from pprint import pprint
```

```python
resp_obj = requests.get("http://gmail.com")
```

```python
resp_obj.history
```

```python
resp_obj.status_code
```

```python
resp_obj.url
```

```python
if resp_obj.history:
    print("Redirect History: \n")
    for resp in resp_obj.history:
        print(resp_obj.status_code, resp_obj.url)
    
    print("\nFinal Destination: \n")
    print(resp_obj.status_code, resp_obj.url)

else:
    print("Server did not redirect.\n")
```
